#### Scrape the website for the info needed

```python
lst=[]
for i in range(1,209):
    url=f"https://www.pewresearch.org/search/immigration/page/{i}/"
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html.parser")
    page=soup.find("ul",{"class":"wp-block-post-template is-layout-flow wp-block-post-template-is-layout-flow"})
    lst.append(page)
lst    
```

#### Narrow down on the features needed to form a dataframe.

```python
l_label=[]
l_time=[]
l_header=[]
l_description=[]
for l in lst:
    articles=l.find_all("article")
    for a in articles:
        label=a.find("span",{"class":"label"}).text.strip()
        l_label.append(label)
        time=a.find("time",{"class":"date"})["datetime"]
        l_time.append(time)
        header=a.find("a").text.strip()
        l_header.append(header)
        description=a.find("div",{"class":"description"}).text.strip()
        l_description.append(description)
        
```

```python
df=pd.DataFrame({"Header":l_header,"Description":l_description,"Time":l_time,"Label":l_label})
df
```
